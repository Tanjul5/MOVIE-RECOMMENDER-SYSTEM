```python
#
import ast
def convert(obj):
  L= []
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L
```

```python
convert(obj)
```

```python
dataset['genres']=dataset['genres'].apply (convert)
```

```python
dataset.head(2)
```

```python
dataset['keywords']=dataset['keywords'].apply (convert)
```

```python
dataset.head(1)
```

```python
obj2=dataset.iloc[0].cast
obj2
```

```python
def convert3 (obj2):
  L=[]
  counter=0
  for i in ast.literal_eval(obj2):
    if counter<3:
      L.append(i['name'])
    counter+=1
  return L
```

```python
convert3 (obj2)
```

```python
dataset['cast']=dataset['cast'].apply (convert3)
```

```python
dataset
```

```python
obj3=dataset.iloc[0].crew
```

```python
#fetch director

def fetch_director(obj3):
    L = []
    for i in ast.literal_eval(obj3):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
dataset['crew']=dataset['crew'].apply(fetch_director)
```

```python
dataset
```

```python
dataset['overview'][0]
```

```python
dataset['overview'] = dataset['overview'].apply(lambda x:x.split())
```

```python
#remove space between words so we can create tags 
dataset['genres']=dataset['genres'].apply (lambda x:[i.replace("","") for i in x])
dataset['keywords']=dataset['keywords'].apply (lambda x:[i.replace("","") for i in x])
dataset['cast']=dataset['cast'].apply (lambda x:[i.replace("","") for i in x])
dataset['crew']=dataset['crew'].apply (lambda x:[i.replace("","") for i in x])
```

```python
dataset['tags']= dataset['overview'] + dataset['keywords'] + dataset['genres'] + dataset['cast'] + dataset['crew']
```

# **VECTORIZATION**

```python
from sklearn.feature_extraction.text import CountVectorizer


cv = CountVectorizer(max_features=5000,stop_words='english')

vector = cv.fit_transform(new_df['tags']).toarray()
```

```python
vector.shape
```

```python
cv.get_feature_names()
```

```python
!pip install nltk
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps=PorterStemmer()
```

```python
def stem(text):
  y=[]
  
  for i in text.split():
    y.append(ps.stem(i))

  return " ".join(y)
```

```python
new_df['tags']= new_df['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity


similarity = cosine_similarity(vector)
```

```python
similarity[0]
```

```python
new_df[new_df['title'] == 'The Lego Movie'].index[0]
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])
```

```python
def recommend(movie):
    index = new_df[new_df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])

    for i in distances[1:6]:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
